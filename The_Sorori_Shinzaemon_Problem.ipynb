```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# [Question 1] Number of rice grains on the 100th day
# Initialise - starting from 1
rice_grain = 1

# Let us create a list to store the cummulative number of rice grains per day
doubled_grains = [rice_grain]

# DOuble the grains and iterate 99 times for a total of 100 days: we'll use a 'for' statement
for _ in range(99):
  rice_grain *= 2
  doubled_grains.append(rice_grain)

# Display the list of doubled_grains and the total on day 100
print(doubled_grains)
print("Total Grains On Day 100: ", doubled_grains[-1])
```

```python
# Line graph to visualise the data
days = list(range(1, 101)) # takes care of 100 days
grains = doubled_grains
plt.figure(figsize=(12, 6))
plt.plot(days, grains)

plt.title('Doubled Grains Over Time')
plt.xlabel('Number of days')
plt.ylabel('Number of grains of rice')
plt.grid(True)

plt.show()
```

```python

# [Question 2] Number of rice grains other than the 100th day
def compute_sorori_shinzaemon(n_days=100):

  """
  A function to calculate the number of grains of rice for each day in the rice anecdote of Sorori Shinzaemon.

  Parameters
  -------
  n_days: int
  Number of days to receive rice (default: 100)

  Returns
  -------
  list_n_grains: list
  List of the number of grains of rice to receive on a certain day (length of list is equal to n_days)
  list_total_grains: list
  List of the total number of grains to receive up to a certain day (length of list is equal to n_days)
  """

# Write code here. pass is a command that means to do nothing.
  list_n_grains = []
  list_total_grains = []

  grains = 1
  total = 0

  # using for statement
  for day in range(n_days):
      list_n_grains.append(grains)
      total += grains
      list_total_grains.append(total)
      grains *= 2 # this doubles the amount of rice grains for the next day

  pass
  return list_n_grains, list_total_grains
n_days = 90 # target days for the course
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)

print(f"Rice grains received on day {n_days}: {list_n_grains[-1]:,}")
print(f"Total rice grains after {n_days} days: {list_total_grains[-1]:,}")

# Visualising the data
plt.figure(figsize=(12, 6))
plt.plot(range(1, n_days + 1), list_n_grains, linestyle='dotted', color='blue', label='Grains/day')
plt.plot(range(1, n_days + 1), list_total_grains, linestyle='solid', color='red', label='Total Grains')

plt.title('Rice grain doubles - Anecdote')
plt.xlabel('Number of days')
plt.ylabel('Number of grains of rice')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# [Question 3] How many people can survive on the rice they receive for how many days?
"""
On average, 1 cup of cooked rice = ~200 grams = ~6,500 grains of rice

A person needs about 2,000 calories per day

Cooked rice (1 cup) provides ~200 calories

Therefore, a person might need about 4.5 cups (900 calories) of rice daily for basic survival (along with other food sources), which equals approximately 30,000 grains of rice per day per person

Assumption for this function:
Number of people = 40
A human needs approximately 30,000 grains of rice per day to survive
"""
def calculate_survival_days_on_rice(total_grains, num_people):
  """
    Calculates how many full days a group of people can survive
    based on the total amount of rice grains available.

    Parameters
    ----------
    total_grains : int
        Total number of grains of rice available.
    num_people : int
        Number of people to feed.

    Returns
    -------
    survival_days : int
        Number of full days the group can survive on the rice.
    """
  # based of research, one person consumes 30000 grains of rice
  grains_per_person_per_day = 30000
  # n_days = 90
  # total_grains = compute_sorori_shinzaemon(n_days)

  # daily group consumption of rice
  daily_group_consumption = num_people * grains_per_person_per_day

  # Calculate the number of full days the group can survive
  survival_days = total_grains // daily_group_consumption

  return survival_days

  # Test
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=90)
total_grains = list_total_grains[-1]
num_people = 40
survival_days = calculate_survival_days_on_rice(total_grains, num_people)
print(f"40 people can survive for {survival_days}")

```
