```python
# 3. Dataset Preparation

from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset with as_frame=True to get DataFrames
iris = load_iris()

# Get feature DataFrame with specified column names
X = iris['data']
X = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

# Get target Series and convert to DataFrame with column name 'Species'
y = pd.DataFrame(iris['target'], columns=['Species'])

# Map target numbers to species names (optional)
y['Species'] = y['Species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Verify the DataFrames
print("Features DataFrame (X):")
print(X.head())
print("\nTarget DataFrame (y):")
print(y.head())
```

```python
# Joining Data
# combining X and y
df = pd.concat([X, y], axis=1)

# verify
print(df.head())
```

```python
# creating diagrams

# 1. Pie Chart for Class Distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
df['Species'].value_counts().plot.pie(
    autopct='%.1f%%',
    colors=['#ff9999','#66b3ff','#99ff99'],
)
plt.title('Iris Species Distribution', fontsize=14)
plt.ylabel('')
plt.show()

# 2. Box Plots for Feature Distribution by Species
plt.figure(figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Species', y=feature, data=df, palette='Pastel1')
    plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()

# 3. Violin Plots Feature Distribution by Species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='Species', y=feature, data=df, palette='Set2', inner='quartile')
    plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()
```

```python
# Checking the relationship between features
# a. 1. Paired Scatter Plots (Colored by Species)
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style='ticks', palette='husl')

# Create pairplot
pair_grid = sns.pairplot(df, hue='Species', height=2.5, markers=['o', 's', 'D'])
pair_grid.fig.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show()
```

```python
# b. Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))

# Calculate correlations
corr_matrix = df.iloc[:, :4].corr()

# Create heatmap
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            vmin=-1, vmax=1,
            linewidths=0.5,
            square=True)

plt.title('Feature Correlation Matrix', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

**Explanation of results in visualized graphs and tables**

Key Insights:

From Pairwise Scatter Plots:

- Clear separation of setosa (blue) in petal dimensions

- Strong linear relationship between petal_length and petal_width

- Versicolor (orange) and virginica (green) overlap in sepal measurements

- Distinct clustering patterns visible in petal feature combinations

From the heatmap

- Very high correlation (0.96) between petal_length and petal_width

- Moderate correlation (0.87) between sepal_length and petal_length

- Negative correlation between sepal_width and petal features

- Weakest correlation (-0.12) between sepal_length and sepal_width

These visualizations reveal that petal measurements are the most discriminative features for classifying iris species, while sepal width shows the least correlation with other features. The high correlation between petal dimensions suggests potential redundancy that could be addressed with dimensionality reduction techniques like PCA.
