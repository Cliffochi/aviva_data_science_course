```python
# in this exercise we will create a function that calculates how much time it takes for chestnuts doubling in volume to fill a certain volume
# Assumptions: Chest nut bun volume = 3.606e-14 cubic km , Solar System = 1.41e40 cubic km, Tokyo Dome = 1.24e-3 cubic km
import math

def minutes_to_fill_solar_systems(initial_chest_nut_volume = 3.606e-14, target_volume=1.41e40, growth_interval = 5):


  # timer in minutes
  time = 0
  multiplier_counter = 0
  volume = initial_chest_nut_volume

  # doubling the volume every 5min
  while volume < target_volume:
    volume *= 2
    time += growth_interval
    multiplier_counter += 1

  print(f"It will take {time} minutes to fill the Solar System with chestnut buns")
  return time, multiplier_counter
```

```python
# Chestnut Bun Growth Simulation in Space

import matplotlib.pyplot as plt
import numpy as np

# Constants (estimated based on research)
initial_volume = 3.606e-14  # average volume of a chestnut bun in cubic km
solar_system_volume = 1.41e40  # rough estimate of solar system volume in km^3

def time_to_fill_volume(initial_volume, max_volume, growth_rate_minutes=5):

    volume_list = [initial_volume]
    minutes_list = [0]

    current_volume = initial_volume
    minutes = 0

    while current_volume < max_volume:
        current_volume *= 2
        minutes += growth_rate_minutes
        volume_list.append(current_volume)
        minutes_list.append(minutes)

    return minutes_list, volume_list

# Test run the function
minutes, volumes = time_to_fill_volume(initial_volume, solar_system_volume)

# Plotting the volume growth
plt.figure(figsize=(12, 6))
plt.plot(minutes, volumes, color='darkorange', linestyle='--', linewidth=2, marker='o')
plt.yscale('log')
plt.title('Chestnut Bun Growth After Sprinkling Baibain', fontsize=14)
plt.xlabel('Time (minutes)', fontsize=12)
plt.ylabel('Volume (km^3)', fontsize=12)
plt.grid(True, which="both", linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Print final result
final_time = minutes[-1]
final_count = len(minutes) - 1
print(f"\n\U0001F4C8 It would take {final_time} minutes ({final_count} doubles) for the chestnut bun to fill the solar system!")
```
