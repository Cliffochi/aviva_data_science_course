```python
# [Problem 1] Creating random numbers
import numpy as np

# set random seet reproducibility
np.random.seed(0)

# distribution parameters
mean =[-3, 0] # Mean vector - specifies the center of the distribution
cov = [[1.0, 0.8],
       [0.8, 1.0]] # Covariance matrix - positive correlation of 0.8 between dimensions)
# generating 500 samples
samples = np.random.multivariate_normal(mean, cov, size=500)

# confirm the shape
print(samples.shape) # expects (500, 2)

# show 10 samples
print("first 10 samples:")
print(samples[:10])
```

```python
# [Problem 2] Visualization using scatter plots
import matplotlib.pyplot as plt

plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Normal Distribution Samples')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
# [Problem 3] Visualization using histograms
# set random seed generator
np.random.seed(0)
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
samples = np.random.multivariate_normal(mean, cov, 500)

# creating the two subplots
plt.figure(figsize=(10, 5))

# first dimension(x-axis values)
plt.subplot(1,2,1)
plt.hist(samples[:, 0], bins=20, color='blue')
plt.title('Histogram of X1')
plt.xlabel('X1')
plt.ylabel('Frequency')
plt.xlim(-6, 2)

# second dimension(y-axis values)
plt.subplot(1,2,2)
plt.hist(samples[:, 1], bins=20, color='red')
plt.title('Histogram of X2')
plt.xlabel('X2')
plt.ylabel('Frequency')
plt.xlim(-6, 2)

plt.tight_layout()
plt.show()
```

```python
# [Problem 4] Adding data
# we will pull data from Q1

# generating Q4 data
mean2 = [0, -3]
data2 = np.random.multivariate_normal(mean2, cov, 500)

# creating the scatter plot
plt.figure(figsize=(10, 8))

# plotting the two datasets with different attributes
scatter1 = plt.scatter(samples[:, 0], samples[:, 1], color='blue', alpha=0.5, label='Dataset 1')
scatter2 = plt.scatter(data2[:, 0], data2[:, 1], color='red', alpha=0.5, label='Dataset 2')

# labeling
plt.title('Comparison of Two 2D Normal Distributions', pad=20)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='gray', linestyle='--', alpha=0.3)
plt.axvline(0, color='gray', linestyle='--', alpha=0.3)
plt.grid(alpha=0.2)
plt.legend()
plt.xlim(-6, 4)
plt.ylim(-6, 4)
```

```python
# [Problem 5] Data Combining
combined_array = np.concatenate((samples.reshape(-1, 1), data2.reshape(-1, 1)), axis=1)

print(combined_array.shape)  # Output: (1000, 2)
print(combined_array[:5])
```

```python
# [Problem 6] Labeling
# Create label array: 0 for first 500 (problem 1), 1 for next 500 (problem 4)
labels = np.concatenate([
    np.zeros(500, dtype=int),
    np.ones(500, dtype=int)
]).reshape(-1, 1)

# Combine into (1000, 3): 2 columns of data + 1 column of labels
labeled_data = np.concatenate([combined_array, labels], axis=1)

# confirm the results
# Check result
print(labeled_data.shape)  # Output: (1000, 3)
print(labeled_data[:5])    # Show first few rows
```
